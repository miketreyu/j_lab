### Vemos el bin() y operadores & y << >>
Está difícil

```python
var1 = 876
print(var1, bin(var1))
var2 = var1<<2
print(var2, bin(var2))
var3 = var1>>2
print(var3, bin(var3))
var4 = 567
print(var4, bin(var4))
var5 = 345
print(var5, bin(var5))
var6 = var4 & var5
print(var6, bin(var6))
```

```python
import random
import string

longitud = 10
caracteres = string.ascii_letters + string.digits
cadena_aleatoria = ''.join(random.choice(caracteres) for i in range(longitud))
print(cadena_aleatoria)
```

## Estructuras condicionales

### Ahora veremos el if:

```python
edad = int(input("Introduce tu edad:"))
if edad >=18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

### Un if más completo

```python
precio = int(input("¿Cuánto te ha costado?"))

if precio < 50:
    categoria = "Barato"
else:
    if precio < 100:
        categoria = "Moderado"
    else:
        categoria = "Caro"
print("El producto es: ", categoria)
```

### Ahora los switch (en python se les llama match) case dentro de las estructuras condicionales

```python
dia = int(input("Introduce el número de día de la semana en que estés:"))
match dia:
    case 1:
        print("Estás a Lunes")
    case 2:
        print("Estás a Martes")
    case 3:
        print("Estás a Miércoles")
    case 4:
        print("Estás a Jueves")
    case 5:
        print("Estás a Viernes")
    case 6:
        print("Estás a Sábado")
    case 7:
        print("Estás a Domingo")
    case _:
        print("Este número no representa ningún día de la semana")
```
