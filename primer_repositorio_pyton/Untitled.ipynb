```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
fscope = open("scope.txt", "r")
scope = fscope.read().splitlines()
fscope.close()
```

```python
print(scope)
```

```python
for activo in scope:

    datos = yf.Ticker(activo).history(period="1y")

    precio = datos['Close']
    volumen = datos['Volume']

    plt.figure(figsize =(12, 6))
    plt.plot(precio)
    plt.title('Precio de ' + activo + 'a lo largo del tiempo')
```
